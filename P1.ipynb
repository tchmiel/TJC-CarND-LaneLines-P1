## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """​weighted_img(img, initial_img, α=0.8, β=1., λ=0.)
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.
# my helper functions
def saveImage(imageFilename, image):  
    RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ensure_dir(imageFilename)
    cv2.imwrite(imageFilename,RGB_img)

debug = False
def debug_plot_image_cmap(image, cmap):
    global debug
    if (debug):
        plt.imshow(image, cmap)
        plt.show() 
        
def debug_plot_image(image):
    global debug
    if (debug):
        plt.imshow(image)
        plt.show() 

def debug_save_image(imageName, image, convertColor=True):
    # debug method to save immediate images
    global debug
    if (debug):
        outDir = "debug_output/"
        baseFilename = imageName + '.jpg'
        imageFilename = os.path.join(outDir, baseFilename)
        print(imageFilename)
        ensure_dir(imageFilename)
        if convertColor:
            RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            cv2.imwrite(imageFilename,RGB_img)
        else:
            cv2.imwrite(imageFilename,image)
       
        
def find_lanes(image) :
    
    # Define some common local variables
    image_y = image.shape[0]
    image_x = image.shape[1]
    
    # convert image to grayscale
    gray_image = grayscale(image)
    #debug_plot_image_cmap(gray_image, 'gray')
    #debug_save_image("Grayscale", gray_image, False)
    
    # Look specificially for yellow lane lanes and combine with gray_image 
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # define range of yellow color in HSV
    yellow_lower = np.array([20,100,100])
    yellow_upper = np.array([30,255,255])
    yellow_mask = cv2.inRange(hsv, yellow_lower, yellow_upper)
    white_mask = cv2.inRange(gray_image, 200, 255)
    both_mask = cv2.bitwise_or(white_mask, yellow_mask)
    mod_gray_image = cv2.bitwise_and(gray_image,both_mask)
    #debug_plot_image(yellow_mask)
    #debug_plot_image(white_mask)
    #debug_plot_image(both_mask)
    #debug_plot_image(mod_gray_image)
    
    # Gaussina smoothing / blurring
    # kernal size, must be odd (3, 5, 7, ...)
    kernel_size = 5
    blur_gray_image = gaussian_blur(mod_gray_image, kernel_size)
    #debug_plot_image_cmap(blur_gray_image, 'gray')
    #debug_save_image("GaussianBlur", blur_gray_image, False)
    
    # apply the canny transform 
    # Note:  recommend a low to high ratio of 1:2 or 1:3 (85/170 or 64/192)
    low_threshold = 64
    high_threshold = 255 - low_threshold
    edges = canny(blur_gray_image, low_threshold, high_threshold)
    debug_plot_image_cmap(edges, 'Greys_r')
    debug_save_image("edges", edges, False)
    
    # Create a masked image
    x_factor = 40
    y_factor = 60
    vertices = np.array([[
        (0,image_y),
        (((image_x/2)- x_factor), (image_y/2)+ y_factor), 
         (((image_x/2) + x_factor), (image_y/2)+ y_factor), 
         (image_x,image_y)]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
    #debug_plot_image(masked_edges)
    #debug_save_image("masked_edges", masked_edges, False)
    
    # Apply the Hough Transform to Edge detected image, return line_image
    rho = 3 
    theta = 1* (np.pi/180) 
    threshold = 15 
    min_line_len = 50 
    max_line_gap = 100 
    line_image = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)
    #debug_plot_image(line_image)
    #debug_save_image("HoughLines", line_image)
    
    # Draw the lines on the original image
    lane_lines = weighted_img(line_image, image)
    #debug_plot_image(lane_lines)
    #debug_save_image("LaneLines", lane_lines)
    
    return lane_lines

def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)
        
def create_subDir (dir, sub_dir):
    new_path = os.path.join(dir, sub_dir)
    ensure_dir(new_path)
    return (new_path)
    
def find_lanes_in_images(dir, output_dir):
    for entry in os.scandir(dir):
        if entry.is_dir():
            continue
        elif entry.is_file():
            filename = entry.path
            image = mpimg.imread(filename)
            lanes_image = find_lanes(image)
            out_filename = os.path.join(output_dir, os.path.basename(filename))
            saveImage(out_filename, lanes_image)
            plt.imshow(lanes_image)
            plt.show()
            continue    
````

#  draw_lines improvement implementation

````python
# Global Variables
g_left_slope = []
g_left_intercept = []
g_right_slope = []
g_right_intercept = []

def reset_draw_line_globals():
    return
    global g_left_slope      # Needed to modify global copy 
    global g_left_intercept  # Needed to modify global copy 
    global g_right_slope     # Needed to modify global copy 
    global g_right_intercept # Needed to modify global copy 
    
    if g_left_slope:
        g_left_slope = []
    if g_left_intercept:
        g_left_intercept = []
    if g_right_slope:
        g_right_slope = []
    if g_right_intercept:
        g_right_intercept = []
    
# source: https://gordoncluster.wordpress.com/2014/02/13/python-numpy-how-to-generate-moving-averages-efficiently-part-2/
def moving_average (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    print (lines)
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
       
    global g_left_slope      # Needed to modify global copy 
    global g_left_intercept  # Needed to modify global copy 
    global g_right_slope     # Needed to modify global copy 
    global g_right_intercept # Needed to modify global copy 
    
    # store a list of the right and right coordinates of each line.
    r_x1 = []
    r_y1 = []
    l_x1 = []
    l_y1 = []
    
    # add a tolerance to avoid lane line jumping
    slope_tolerance = 0.30
    
    for line in lines:
        for x1,y1,x2,y2 in line: 
            # avoid vertical line, slope undefined
            if (x2 == x1):
                continue
            
            # calculate slope
            line_slope = (y2 - y1) / (x2 - x1)
            
            # watch slope - slope in reversed because of image's upper left coordinate is (0,0)
            if line_slope > slope_tolerance:
                # right lane line
                r_x1.append(x1)
                r_y1.append(y1)
                r_x1.append(x2)
                r_y1.append(y2)
                    
            elif line_slope < -slope_tolerance:
                #print (x1,x2, y1, y2, line_slope)
                l_x1.append(x1)
                l_y1.append(y1)
                l_x1.append(x2)
                l_y1.append(y2)
    
    #test for no points to make up a right lane
    if (len(r_x1) == 0):
        return
    
    #test for no points to make up a left lane
    if (len(l_x1) == 0):
        return
    
    
    # Perform Linear regression on stored coordinates, to determine slope of each line.
    from scipy.stats import linregress
    (r_slope, r_intercept, r_rvalue, r_pvalue, r_stderr) = linregress(r_x1, r_y1)
    (l_slope, l_intercept, l_rvalue, l_pvalue, l_stderr) = linregress(l_x1, l_y1)
    
    # store of the slope to perform a moving average later
    g_left_slope = np.append (g_left_slope, l_slope)
    g_left_intercept =  np.append (g_left_intercept, l_intercept)
    g_right_slope = np.append (g_right_slope, r_slope)
    g_right_intercept =  np.append (g_right_intercept, r_intercept)
    
    # create an average line base on average slope and intercept
    #    fix the length of the line lane to be the bounding polygon
    y1 = img.shape[0]
    y2 = int(y1 * 0.6) # use bottom 40% of the image 
    
    # create a moving average to miminize jumpiness between frames
    window = 3
    # iteration
    if (len(g_left_slope) > window):
        # after you have ten frames, start using the moving average
        rollavg_left_slope = moving_average(g_left_slope, window)
        rollavg_left_intercept = moving_average(g_left_intercept, window)
        rollavg_right_slope = moving_average(g_right_slope, window)
        rollavg_right_intercept = moving_average(g_right_intercept, window)
        
        left_x1 = int((y1 - rollavg_left_intercept[-1]) / rollavg_left_slope[-1])
        left_x2 = int((y2 - rollavg_left_intercept[-1]) / rollavg_left_slope[-1])
        left_y1 = int(y1)
        left_y2 = int(y2)
   
        right_x1 = int((y1 - rollavg_right_intercept[-1]) / rollavg_right_slope[-1])
        right_x2 = int((y2 - rollavg_right_intercept[-1]) / rollavg_right_slope[-1])
        right_y1 = int(y1)
        right_y2 = int(y2)
    
    else:
        # less than moving average window / use the curent slopes for this frame
        left_x1 = int((y1 - l_intercept) / l_slope)
        left_x2 = int((y2 - l_intercept) / l_slope)
        left_y1 = int(y1)
        left_y2 = int(y2)
    
        right_x1 = int((y1 - r_intercept) / r_slope)
        right_x2 = int((y2 - r_intercept) / r_slope)
        right_y1 = int(y1)
        right_y2 = int(y2)
        
    cv2.line(img, (left_x1, left_y1), (left_x2, left_y2), color, thickness)
    cv2.line(img, (right_x1, right_y1), (right_x2, right_y2), color, thickness)
    
````
